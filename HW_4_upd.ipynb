Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
### YOUR CODE HERE
import numpy as np
import sympy as sym
import math
from scipy.misc import derivative

x = sym.Symbol('x')

# 1-я производная
df1 = sym.diff(x**5 + 4 * sym.sin(2 * x) + sym.cos(3 * x + 3), x)
df1
```

```python
def f1(x):
    return x**5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)
print(f'1-я производная:{derivative(f, 1.0, dx=1e-6, n=1)}')
```

```python
# 2-я производная
def f2(x):
    return x**5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)
print(f'2-я производная:{derivative(f, 1.0, dx=1e-6, n=2)}')

# Это локальный максимум. Функция строго вогнутая.
```

### Задание 2

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
### YOUR CODE HERE
import matplotlib.pyplot as plt

# df_ = sym.diff(sym.sin(2 * x + 1)**5, x)
# df_
```

```python
def f3(x):
    return (np.sin(2 * x + 1))**5
print(f'1-я производная:{derivative(f, 1.0, dx=1e-6, n=1)}')
```

```python
x = np.linspace(-5, 5)
fx = f3(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```

```python
def f4(x):
    return (np.sin(2 * x + 1))**5
# x = np.arange(-5,5)
print(f'2-я производная:{derivative(f, 1.0, dx=1e-6, n=2)}')
```

```python
x = np.linspace(-5, 5)
fx = f4(x)
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
# f2x
```
